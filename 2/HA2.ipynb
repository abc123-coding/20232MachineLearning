  3. First, train and test SVM using linear kernel

  4. Second, train and test SVM using RBF kernel

  5. Hyperparemter tuning by using Grid Search (i.e., from sklearn.model_selection import GridSearchCV)

```python
# TO-DO 2. Use "StandardScaler" function (provided in HA1_part1) to map all points to have mean=0 and std=1

import pandas as pd
from sklearn.preprocessing import StandardScaler

df_X = pd.DataFrame(X, columns = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"])
X_normalized = StandardScaler().fit_transform(df_X.values)
X_normalized
```

```python
# TO-DO 1. Split the dataset into training and testing sets (training:testing=75:25)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
#  TO-DO 3. First, train and test SVM using linear kernel
from sklearn.svm import LinearSVC

classifier = LinearSVC(C=10000, loss='hinge')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
# TO-DO 3. Report the accuracy
from sklearn.metrics import accuracy_score

print("accuracy with Linear Kernel: {}".format(accuracy_score(y_test, y_pred)))
```

```python
# TO-DO 3. By using 5-fold cross validation, again report accuracies from each fold and the mean accuracy after taking the mean on five accuracies.
from sklearn.model_selection import cross_val_score

scores = cross_val_score(classifier, X_train, y_train, cv=5)
print('cross-val-score: {}'.format(scores))
print('cross-val-score.mean: {}'.format(scores.mean()))
```

```python
# TO-DO 4. Second, train and test SVM using RBF kernel
#  use C=1, gamma=0.01
#  Report the accuracy
#  After that, by using 5-fold cross validation, again report accuracies from each fold and the mean accuracy after taking the mean on five accuracies.

from sklearn.svm import SVC

classifier_w_rbf = SVC(kernel = 'rbf', C=1, gamma = 0.01)
classifier_w_rbf.fit(X_train, y_train)
y_pred_rbf = classifier.predict(X_test)
print("accuracy with RBF kernel: {}".format(accuracy_score(y_test, y_pred_rbf)))

scores = cross_val_score(classifier_w_rbf, X_train, y_train, cv=5)
print('cross-val-score: {}'.format(scores))
print('cross-val-score.mean: {}'.format(scores.mean()))
```

```python
# TO-DO 5. Hyperparemter tuning by using Grid Search (i.e., from sklearn.model_selection import GridSearchCV)
#   Parameter search space must include three kernels (['rbf', 'poly', 'sigmoid'], C ([0.1,1, 10, 100]) and gamma ([1,0.1,0.01,0.001]))
#   Report best parameters and accuracy based on them

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

param_grid = {
              "kernel":['rbf', 'poly', 'sigmoid'],
              'C':[0.1, 1, 10, 100],
              "gamma":[1, 0.1, 0.01, 0.001]
              }

svm = SVC()
gscv = GridSearchCV (estimator = svm, param_grid=param_grid, cv = 5)
gscv.fit(X_train, y_train)

print("best parameters: ", gscv.best_params_)
print("Best accuracy: ", gscv.best_score_)
```

A. The best parameters are given with "rbf" kernel because it is commonly used for SVM and tends to work well for many cases. It is capable of computing weights considering the distance of neighboring points.
